```python
#cek value yang null berdasarkan kolom
for col in data.columns:
    print(col, data[col].isnull().sum())
```

```python
#cek value yang NaN berdasarkan kolom
for col in data.columns:
    print(col, data[col].isna().sum())
```

```python
#menghapus baris berdasarkan NaN value
data2 = data.dropna()
```

```python
#mengecek kembali NaN value berdasarkan kolom
for col in data2.columns:
    print(col, data2[col].isna().sum())
```

```python
#cek jumlah baris dan kolom setelah dilakukan penghapusan baris 
data2.shape
```

```python
#mengambil atribut products
products = data2.loc[:,['products']]
```

```python
products
```

```python
#mengambil 1000 baris dari variable products
products2 = products[0:1000]
```

```python
#mengubah caracter "|" menjadai "," (komma), dan ditampung oleh atribut baru
products2['Changes_line'] = products2['products'].apply(lambda x: 
                                                    [re.sub(r"[\|]", ",", word) for word in x.split(",")])
```
